```python
import pandas as pd
import csv
```

```python
def csv_to_dict(path):
    """Returns a doctionary from the result csv file"""
    table = pd.read_csv(path, sep = ';')
    table = table[['image', 'used_liter']]
    table = table.set_index('image')['used_liter']
    return table.to_dict()
```

```python
csv_to_dict('data/MQ_digital.csv')
```
